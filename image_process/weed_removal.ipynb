```python
#基本导入
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
path_raw = 'C:\\Users\\xianyu\\GraduationProject\\UAV_YUNNAN_DATA\\0-raw-img\\batch-1\\P061301.tif'
path_predict_bin = 'C:\\Users\\xianyu\\GraduationProject\\UAV_YUNNAN_DATA\\3-predict_result\\P061301_result.tif'

# img_raw = cv.imread(path_raw)[3500:4000, 6000:6500]
# img_raw = cv.GaussianBlur(img_raw, (3, 3), 1)

img_predict_bin = cv.imread(path_predict_bin, 0)[3500:4000, 6000:6500]

plt.imshow(img_predict_bin, cmap='gray')
```

```python
# 开运算
plt.figure(figsize=(8, 8), dpi=300)
kernel3 = np.ones((3, 3), np.uint8)
kernel5 = np.ones((5, 5), np.uint8)

plt.subplot(2, 2, 1)
plt.title('original')
plt.imshow(img_predict_bin, cmap='gray')

plt.subplot(2, 2, 2)
open_m3 = cv.morphologyEx(img_predict_bin, cv.MORPH_OPEN, kernel5)
plt.title('5*5')
plt.imshow(open_m3, cmap='gray')

plt.subplot(2, 2, 3)
open_m1 = cv.morphologyEx(img_predict_bin, cv.MORPH_OPEN, kernel3)
plt.title('3*3')
plt.imshow(open_m1, cmap='gray')

plt.subplot(2, 2, 4)
open_m2 = cv.morphologyEx(img_predict_bin, cv.MORPH_OPEN, kernel3, iterations=2)
plt.title('3*3, iter=2')
plt.imshow(open_m2, cmap='gray')
```

```python
# 开运算
kernel = np.ones((5, 5), np.uint8)
open_m = cv.morphologyEx(img_predict_bin, cv.MORPH_OPEN, kernel)
plt.imshow(open_m, cmap='gray')
```
